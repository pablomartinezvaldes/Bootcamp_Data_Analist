## 16 - Normalice una matriz aleatoria de 10x10.

`pista: (x - mean) / std`

```python
# Crear una matriz aleatoria de 10x10
matriz_10x10 = np.random.random((10, 10))

# Normalizar la matriz
media = np.mean(matriz_10x10)
desviacion_estandar = np.std(matriz_10x10)
matriz_normalizada = (matriz_10x10 - media) / desviacion_estandar

print(matriz_normalizada)
```

## 21 - Extraiga la parte entera de una matriz aleatoria de números positivos usando 4 métodos diferentes.

`pista: %, np.floor, astype, np.trunc`

```python
# Crear una matriz aleatoria de números positivos
matriz_aleatoria = np.random.random((5, 5)) * 10
print("Matriz original:")
print(matriz_aleatoria)

# Método 1: Usando el operador de módulo
parte_entera_modulo = matriz_aleatoria // 1
print("\nParte entera usando operador de módulo:")
print(parte_entera_modulo)

# Método 2: Usando np.floor
parte_entera_floor = np.floor(matriz_aleatoria)
print("\nParte entera usando np.floor:")
print(parte_entera_floor)

# Método 3: Usando astype(int)
parte_entera_astype = matriz_aleatoria.astype(int)
print("\nParte entera usando astype(int):")
print(parte_entera_astype)

# Método 4: Usando np.trunc
parte_entera_trunc = np.trunc(matriz_aleatoria)
print("\nParte entera usando np.trunc:")
print(parte_entera_trunc)
```

## 29 - ¿Cómo encontrar el valor más cercano, a un escalar dado, en un vector?

`pista: argmin`


```python
# Crear un vector de ejemplo
vector_ejemplo = np.random.random(10) * 100
print("Vector de ejemplo:", vector_ejemplo)

# Escalar dado
escalar_dado = 50

# Encontrar el valor más cercano
valor_mas_cercano = vector_ejemplo[np.abs(vector_ejemplo - escalar_dado).argmin()]
print(f"El valor más cercano a {escalar_dado} es: {valor_mas_cercano}")
```

## 31 - ¿Cuál es el equivalente de `enumerate` para matrices en `numpy`?

`pista: np.ndenumerate, np.ndindex`

```python
# Crear una matriz de ejemplo
matriz_ejemplo = np.random.random((3, 3))
print("Matriz original:")
print(matriz_ejemplo)

# Usar np.ndenumerate para enumerar elementos
print("\nUsando np.ndenumerate:")
for indice, valor in np.ndenumerate(matriz_ejemplo):
    print(indice, valor)

# Usar np.ndindex para enumerar índices
print("\nUsando np.ndindex:")
for indice in np.ndindex(matriz_ejemplo.shape):
    print(indice, matriz_ejemplo[indice])
```

## 32 - Restar la media de cada fila a una matriz.

`pista: mean(axis=, keepdims=)`


```python
# Crear una matriz de ejemplo
matriz_ejemplo = np.random.random((5, 5))
print("Matriz original:")
print(matriz_ejemplo)

# Restar la media de cada fila
media_filas = matriz_ejemplo.mean(axis=1, keepdims=True)
matriz_resta_media = matriz_ejemplo - media_filas
print("\nMatriz después de restar la media de cada fila:")
print(matriz_resta_media)
```

## 33 - ¿Cómo ordenar una matriz por la enésima columna?

`pista: argsort`

```python
# Crear una matriz de ejemplo
matriz_ejemplo = np.random.random((5, 5))
print("Matriz original:")
print(matriz_ejemplo)

# Ordenar la matriz por la enésima columna (por ejemplo, la tercera columna, índice 2)
n = 2
matriz_ordenada = matriz_ejemplo[matriz_ejemplo[:, n].argsort()]
print(f"\nMatriz ordenada por la columna {n}:")
print(matriz_ordenada)
```

## 34 - ¿Cómo saber si una matriz 2D tiene columnas nulas?

`pista: any, ~`


```python
# Crear una matriz de ejemplo
matriz_ejemplo = np.random.random((5, 5))
print("Matriz original:")
print(matriz_ejemplo)

# Verificar si alguna columna es completamente nula
columnas_nulas = ~matriz_ejemplo.any(axis=0)
print(f"\n¿La matriz tiene columnas nulas?: {columnas_nulas.any()}")
print(f"Columnas nulas: {columnas_nulas}")
```

## 35 - Encuentre el valor más cercano de un valor dado en una matriz.

`pista: np.abs, argmin, flat`


```python
# Crear una matriz de ejemplo
matriz_ejemplo = np.random.random((5, 5)) * 100
print("Matriz de ejemplo:")
print(matriz_ejemplo)

# Valor dado
valor_dado = 50

# Encontrar el valor más cercano
valor_mas_cercano = matriz_ejemplo.flat[np.abs(matriz_ejemplo - valor_dado).argmin()]
print(f"El valor más cercano a {valor_dado} es: {valor_mas_cercano}")
```
